## Imports & Defs



```python
from xopt import Xopt
import os
from concurrent.futures import ProcessPoolExecutor
import numpy as np
import json
from impact import Impact
from distgen import Generator
from matplotlib import pyplot as plt

# Define sequence for simulations
def in1(x,param):
    d2 = SETTINGS0.copy()
    d2[param] = x
    return d2
    
def run1(x, param):
    inputs = in1(x,param)
    res = X.evaluate(inputs)
    return res

def scan1(param, xlist, func):
    d = {}
    d['param'] = param
    d['x'] = xlist.tolist()
    d['reference'] = SETTINGS0.copy()
    with ProcessPoolExecutor() as executor:
        results = list(executor.map(func, xlist))
    d['inputs'] = [in1(x,param) for x in xlist]
    d['outputs'] = results
    
    return d    
```
